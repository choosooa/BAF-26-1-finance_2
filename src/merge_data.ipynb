```python
import pandas as pd
import numpy as np

DATA_DIR = '../datas/Finance_raw/'
START_DATE = '2006-01-01'
```

## 1. 일별 데이터 로드 (FRED)

```python
# OilPrice (일별)
oilprice = pd.read_csv(DATA_DIR + 'FRED_OilPrice.csv', parse_dates=['observation_date'])
oilprice = oilprice.rename(columns={'observation_date': 'date', 'DCOILWTICO': 'OilPrice'})
oilprice = oilprice[oilprice['date'] >= START_DATE][['date', 'OilPrice']].copy().reset_index(drop=True)
print('OilPrice:', oilprice.shape, oilprice['date'].min(), '~', oilprice['date'].max())
```

```python
# DollarIndex (일별)
dollarindex = pd.read_csv(DATA_DIR + 'FRED_DollarIndex.csv', parse_dates=['observation_date'])
dollarindex = dollarindex.rename(columns={'observation_date': 'date', 'DTWEXBGS': 'DollarIndex'})
dollarindex = dollarindex[dollarindex['date'] >= START_DATE][['date', 'DollarIndex']].copy().reset_index(drop=True)
print('DollarIndex:', dollarindex.shape, dollarindex['date'].min(), '~', dollarindex['date'].max())
```

```python
# VIX (일별)
vix = pd.read_csv(DATA_DIR + 'FRED_VIX.csv', parse_dates=['observation_date'])
vix = vix.rename(columns={'observation_date': 'date', 'VIXCLS': 'VIX'})
vix = vix[vix['date'] >= START_DATE][['date', 'VIX']].copy().reset_index(drop=True)
print('VIX:', vix.shape, vix['date'].min(), '~', vix['date'].max())
```

```python
# TermSpread (일별)
termspread = pd.read_csv(DATA_DIR + 'FRED_TermSpread.csv', parse_dates=['observation_date'])
termspread = termspread.rename(columns={'observation_date': 'date', 'T10Y2Y': 'TermSpread'})
termspread = termspread[termspread['date'] >= START_DATE][['date', 'TermSpread']].copy().reset_index(drop=True)
print('TermSpread:', termspread.shape, termspread['date'].min(), '~', termspread['date'].max())
```

```python
# TreasuryYield (일별)
treasuryyield = pd.read_csv(DATA_DIR + 'FRED_TreasuryYield.csv', parse_dates=['observation_date'])
treasuryyield = treasuryyield.rename(columns={'observation_date': 'date', 'DGS10': 'TreasuryYield'})
treasuryyield = treasuryyield[treasuryyield['date'] >= START_DATE][['date', 'TreasuryYield']].copy().reset_index(drop=True)
print('TreasuryYield:', treasuryyield.shape, treasuryyield['date'].min(), '~', treasuryyield['date'].max())
```

## 2. 월별 데이터 로드 (FRED)

```python
# RealInterestRate (월별)
realinterestrate = pd.read_csv(DATA_DIR + 'FRED_RealInterestRate_month.csv', parse_dates=['observation_date'])
realinterestrate = realinterestrate.rename(columns={'observation_date': 'date', 'REAINTRATREARAT10Y': 'RealInterestRate'})
realinterestrate = realinterestrate[realinterestrate['date'] >= START_DATE][['date', 'RealInterestRate']].copy().reset_index(drop=True)
print('RealInterestRate:', realinterestrate.shape, realinterestrate['date'].min(), '~', realinterestrate['date'].max())
```

```python
# CPI (월별)
cpi = pd.read_csv(DATA_DIR + 'FRED_CPI_month.csv', parse_dates=['observation_date'])
cpi = cpi.rename(columns={'observation_date': 'date', 'CPIAUCSL': 'CPI'})
cpi = cpi[cpi['date'] >= START_DATE][['date', 'CPI']].copy().reset_index(drop=True)
print('CPI:', cpi.shape, cpi['date'].min(), '~', cpi['date'].max())
```

```python
# IndustryProduction (월별)
industryproduction = pd.read_csv(DATA_DIR + 'FRED_IndustryProduction_month.csv', parse_dates=['observation_date'])
industryproduction = industryproduction.rename(columns={'observation_date': 'date', 'INDPRO': 'IndustryProduction'})
industryproduction = industryproduction[industryproduction['date'] >= START_DATE][['date', 'IndustryProduction']].copy().reset_index(drop=True)
print('IndustryProduction:', industryproduction.shape, industryproduction['date'].min(), '~', industryproduction['date'].max())
```

```python
# CPE (월별) - FRED_CPE_month.csv
cpe = pd.read_csv(DATA_DIR + 'FRED_CPE_month.csv', parse_dates=['observation_date'])
cpe = cpe.rename(columns={'observation_date': 'date', 'PCEC96': 'CPE'})
cpe = cpe[cpe['date'] >= START_DATE][['date', 'CPE']].copy().reset_index(drop=True)
print('CPE:', cpe.shape, cpe['date'].min(), '~', cpe['date'].max())
```

```python
# FedFundsRate (월별)
fedfundsrate = pd.read_csv(DATA_DIR + 'FRED_FedFundsRate.csv', parse_dates=['observation_date'])
fedfundsrate = fedfundsrate.rename(columns={'observation_date': 'date', 'FEDFUNDS': 'FedFundsRate'})
fedfundsrate = fedfundsrate[fedfundsrate['date'] >= START_DATE][['date', 'FedFundsRate']].copy().reset_index(drop=True)
print('FedFundsRate:', fedfundsrate.shape, fedfundsrate['date'].min(), '~', fedfundsrate['date'].max())
```

## 3. 주별 데이터 로드 (EIA)

```python
# OilInventories (주별) - 상위 4행 메타데이터 + 1행 헤더 스킵
oilinventories = pd.read_csv(DATA_DIR + 'eia_OilInventories_week.csv', skiprows=5, names=['date', 'OilInventories'])
oilinventories['date'] = pd.to_datetime(oilinventories['date'], format='%m/%d/%Y')
oilinventories['OilInventories'] = pd.to_numeric(oilinventories['OilInventories'], errors='coerce')
oilinventories = oilinventories[oilinventories['date'] >= START_DATE][['date', 'OilInventories']].copy().reset_index(drop=True)
print('OilInventories:', oilinventories.shape, oilinventories['date'].min(), '~', oilinventories['date'].max())
```

```python
# OilProduction (주별) - 상위 4행 메타데이터 + 1행 헤더 스킵
oilproduction = pd.read_csv(DATA_DIR + 'eia_OilProduction_week.csv', skiprows=5, names=['date', 'OilProduction'])
oilproduction['date'] = pd.to_datetime(oilproduction['date'], format='%m/%d/%Y')
oilproduction['OilProduction'] = pd.to_numeric(oilproduction['OilProduction'], errors='coerce')
oilproduction = oilproduction[oilproduction['date'] >= START_DATE][['date', 'OilProduction']].copy().reset_index(drop=True)
print('OilProduction:', oilproduction.shape, oilproduction['date'].min(), '~', oilproduction['date'].max())
```

## 4. OPEC 생산량 월별 데이터 로드 (EIA, wide format)

```python
# OPECProduction (월별) - wide format
# INTL.53-1-XXX-TBPD.M: 국가별 'Total petroleum and other liquids' 행 합산
opec_raw = pd.read_csv(DATA_DIR + 'eia_OPECProduction_month.csv', header=1)

total_rows = opec_raw[opec_raw['API'].str.startswith('INTL.53-1-', na=False)]
date_cols = [c for c in opec_raw.columns if c not in ['API', 'Unnamed: 1']]

opec_sum = total_rows[date_cols].apply(pd.to_numeric, errors='coerce').sum(axis=0)
opec_total = opec_sum.reset_index()
opec_total.columns = ['month_str', 'OPECProduction']
opec_total['date'] = pd.to_datetime(opec_total['month_str'], format='%b %Y')
opec_total = opec_total[opec_total['date'] >= START_DATE][['date', 'OPECProduction']].copy().reset_index(drop=True)

# 합산 결과 0인 행 → 데이터 없음으로 처리
opec_total.loc[opec_total['OPECProduction'] == 0, 'OPECProduction'] = np.nan

print('OPECProduction:', opec_total.shape, opec_total['date'].min(), '~', opec_total['date'].max())
opec_total.head()
```

## 5. 외부 조인(Outer Join)으로 병합
- 일별 데이터: 해당 날짜에만 값 존재
- 월별/주별 데이터: 해당 날짜에만 값 존재, 나머지는 NaN

```python
dfs = [
    oilprice,
    realinterestrate,
    cpi,
    dollarindex,
    vix,
    industryproduction,
    cpe,
    oilinventories,
    opec_total,
    oilproduction,
    termspread,
    treasuryyield,
    fedfundsrate,
]

merged = dfs[0]
for df in dfs[1:]:
    merged = pd.merge(merged, df, on='date', how='outer')

merged = merged.sort_values('date').reset_index(drop=True)

# 컬럼 순서 지정
col_order = [
    'date', 'OilPrice', 'RealInterestRate', 'CPI', 'DollarIndex', 'VIX',
    'IndustryProduction', 'CPE', 'OilInventories', 'OPECProduction',
    'OilProduction', 'TermSpread', 'TreasuryYield', 'FedFundsRate'
]
merged = merged[col_order]

print('Merged shape:', merged.shape)
print('Date range:', merged['date'].min(), '~', merged['date'].max())
print()
print('결측치 현황:')
print(merged.isnull().sum())
merged.head(10)
```

## 6. 저장

```python
output_path = DATA_DIR + 'F_merged.csv'
merged.to_csv(output_path, index=False)
print(f'저장 완료: {output_path}')
print(f'Shape: {merged.shape}')
merged.tail()
```
