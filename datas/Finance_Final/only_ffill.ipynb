```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("비어플/F_merged (1).csv")
df
```

```python
# =========================
# 주간 / 월간 변수만 결측치 forward fill
# =========================

weekly_vars = [
    "OilInventories", "OilProduction"
]

monthly_vars = [
    "RealInterestRate", "CPI", "IndustryProduction",
    "CPE", "OPECProduction", "FedFundsRate"
]

# 날짜 정렬 (ffill 전에 필수)
df = df.sort_values("date").reset_index(drop=True)

# 주간 / 월간 변수만 forward fill
df[weekly_vars + monthly_vars] = df[weekly_vars + monthly_vars].ffill()

# 확인
print(df[weekly_vars + monthly_vars].head())

# 저장
# =========================
df.to_csv("model_data_only_ffill.csv", index=False)
print("\n저장 완료: model_data_only_ffill.csv")
```
